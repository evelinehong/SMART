```python
import json
```

```python
with open('../../generalization_data_test.json', encoding="utf-8") as json_file:
   all_data_test = json.load(json_file)
```

```python
with open('./data/new_test23k_processed.json', encoding="utf-8") as json_file:
   data_test = json.load(json_file)
```

```python
from collections import defaultdict
data_by_type = defaultdict(list)

for x in data_test:
    for y in all_data_test:
        if y['id']==x['id']:
            x['type']=y['type']
            data_by_type[x['type']].append(x)
            break
```

```python
with open('pg_seq_norm_True_test.json', encoding="utf-8") as json_file:
   test_results = json.load(json_file)
```

```python
results = {}
for result in test_results:
    results[result['index']] = result
```

```python
for typ, ls in data_by_type.items():
    print("TYPE:")
    print(typ)
    
    value_ac = 0
    eval_total = 0
    
    #print(x)
    for x in ls:
        try:
            if abs(float(results[x['id']]['gen_ans']) - float(results[x['id']]['ans'])) < 1e-5:
                value_ac+=1
        except:
            pass
        eval_total+=1
    print(f"{value_ac/eval_total}")
```
